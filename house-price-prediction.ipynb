```python
import pandas as pd
```

```python
ds = pd.read_csv("/kaggle/input/house-price/house_prices.csv")
ds.head(3)
```

```python
ds.isnull().sum().sort_values(ascending = False)
```

```python
ds.drop(columns=['Super Area', 'Plot Area', 'Dimensions', 'Society', 'overlooking', 'facing', 'Car Parking'], inplace=True)
```

```python
ds.drop(columns = ['Description','Title'], inplace = True)
```

```python
ds.isnull().sum()
```

```python
ds['Carpet Area']= ds['Carpet Area'].str.extract('(\d+)').astype(float)
```

```python
ds['Floor'] = ds['Floor'].str.extract('(\d+)').astype(float)
```

```python
ds['Bathroom'] = ds['Bathroom'].str.extract('(\d+)').astype(float)
```

```python
ds['Balcony'] = ds['Balcony'].str.extract('(\d+)').astype(float)
```

```python
ds['Price (in rupees)'].fillna(ds['Price (in rupees)'].median(), inplace= True)
ds['Carpet Area'].fillna(ds['Carpet Area'].median(), inplace =True)
ds['Bathroom'].fillna(ds['Bathroom'].median(), inplace = True)
ds['Balcony'].fillna(0, inplace  = True)
```

```python
ds['Status'].fillna('Unknown', inplace = True)
ds['Floor'].fillna('Unknown', inplace = True)
ds['Transaction'].fillna('Unknown', inplace = True)
ds['Furnishing'].fillna('Unknown', inplace = True)
ds['Ownership'].fillna('Unknown', inplace = True)
```

```python
ds.isnull().sum()
```

```python
ds = pd.get_dummies(ds, drop_first = True)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x = ds.drop('Price (in rupees)', axis = 1)
y = ds['Price (in rupees)']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 42)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
lr = LinearRegression()
lr.fit(x_train, y_train)
```

```python
lr.score(x_test, y_test)*100
```

```python
pred_y = lr.predict(x_test)
```

```python
rmse = mean_squared_error(y_test, pred_y, squared = False)
print("test rmse", rmse)
```


# let's compareMultiple models


```python
ds.dropna(subset = ['Price (in rupees)'], inplace = True)
```

```python
ds_encoded = pd.get_dummies(ds, drop_first = True)
```

```python
x = ds_encoded.drop('Price (in rupees)', axis = 1)
y = ds_encoded['Price (in rupees)']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size = 0.2, random_state = 42)
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
ss = StandardScaler()
x_train_scaled = ss.fit_transform(x_train)
x_test_scaled = ss.transform(x_test)
```

```python
lr = LinearRegression()
lr.fit(x_train_scaled, y_train)
pred_y = lr.predict(x_test_scaled)
mean_sq_err = mean_squared_error(y_test, pred_y, squared = False)
r2 = r2_score(y_test, pred_y)
```

```python
print("ðŸ”¹ Linear Regression â†’ RMSE:", round(mean_sq_err), "| RÂ²:", round(mean_sq_err, 3))
```

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
dtr = DecisionTreeRegressor(random_state = 42)
dtr.fit(x_train, y_train)
dt_pred = dtr.predict(x_test)
print("mean_square_error: ", mean_squared_error(y_test, dt_pred))
print("r2_score: ", r2_score(y_test, dt_pred))
```

```python
#optimized version of decision tree
for depth in [2,5,10, 20,None]:
    model = DecisionTreeRegressor(max_depth = depth, random_state = 42)
    model.fit(x_train, y_train)
    y_pred = model.predict(x_test)
    mse = mean_squared_error(y_test, y_pred)
    print(f"depth = {depth} --> MSE = {mse:.2f}")
```
